```python
# ===========================[ Demo / Usage ]========================

from datetime import date

# Setup countries
sa = Country("South Africa", currency="ZAR")
na = Country("Namibia",     currency="NAD")
```

```python

# Experiences
kruger = SafariExperience("Kruger Sunrise Safari", sa, base_price=2500.0, capacity=30)
cape   = BeachExperience("Cape Town Beach Day",    sa, base_price=1200.0, capacity=50)  # unused in totals
dune   = AdventureExperience("Sossusvlei Dune Hike", na, base_price=1800.0, capacity=20)
```

```python

# Availability
kruger.set_availability(date(2025, 7, 15), 10)
dune.set_availability(date(2025, 9, 23), 15)
```

```python

# Package
pkg = Package("Safari & Sand Dunes Explorer")
pkg.add_item(PackageItem(kruger, date(2025, 7, 15), nights=1))
pkg.add_item(PackageItem(dune,   date(2025, 9, 23), nights=1, discount=0.05))
```

```python

# Group
group = TouristGroup("Team Ubuntu", size=6, per_person_budget=10_000.0,
                     preferences=["safari", "adventure"])
```

```python

# Booking flow
bm = BookingManager()
print("\nâ€” Booking flow â€”")
try:
    quote = bm.preview_cost(pkg, group)
    print("Preview (no seats reserved) â†’", round(quote, 2))

    booking = bm.confirm(pkg, group)
    print("Confirm â†’ status:", booking.status)
    print("Total due (per-person Ã— group size) â†’", booking.total_cost())

    print("Remaining seats (Kruger 2025-07-15) â†’",
          kruger.available_capacity(date(2025, 7, 15)))

    bm.cancel(booking)
    print("Cancel â†’ status:", booking.status)
    print("Restored seats (Kruger 2025-07-15) â†’",
          kruger.available_capacity(date(2025, 7, 15)))

except (ValidationError, AvailabilityError, BookingError) as e:
    print("Error:", e)
```
